```python
import pandas as pd
import seaborn as sns
from sqlalchemy import engine
from pandasql import sqldf

# Función lambda que hace escribir funciones en SQL mucho más fácil
pysqldf = lambda q: sqldf(q, globals())
```

```python
df = pd.read_csv('/content/datos_google_ads.csv')
```

```python
df
```

```python
df.shape
```

```python
df.describe()
```

```python
#Busqueda de duplicados
df.duplicated('Keyword').sum()
```

```python
#Eliminar datos nulos en columna Competition (indexed value)
df = df.dropna(subset=['Competition (indexed value)'])
```

```python
#Eliminar datos nulos en columna Top of page bid (low range)
df = df.dropna(subset=['Top of page bid (low range)'])
```

```python
df
```

```python
ax = df.plot.hist(column=["Competition (indexed value)"], by='Competition', bins=10, figsize=(8, 12))
```

```python
#Suma de las busquedas promedio para cada nivel de competencia
sns.barplot(data=df, x="Competition", y="Avg. monthly searches", errorbar=None, estimator=sum)
```

Correr SQL en Python

```python
# Renombrar columnas para que el AQL no rompa.
df = df.rename(columns={"Avg. monthly searches": "avg_searches", "Competition (indexed value)": "competition_index"})
```

```python
# Sentencia SQL
q = """
SELECT
Keyword,
avg_searches,
Competition,
competition_index
FROM df
WHERE avg_searches >= 500
"""

df_export = pysqldf(q)
```

```python
#Exportar la salida del select al archivo datos_google_ads_procesados.csv
df_export.to_csv("datos_google_ads_procesados.csv")
```
